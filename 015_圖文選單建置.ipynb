```python
'''

載入前面的圖文選單設定，

並要求line_bot_api將圖文選單上傳至Line
    

'''

from linebot.models import RichMenu
import requests
import json
#讀取圖文選單設定檔
menuJson=json.loads(menuRawData)
#用Line_bot_api的create_rich_menu訪法 創造圖文選單id
#rich_menu : RichMenu.new_from_json_dict >>> 可以從json裡面生成RichMenu物建
lineRichMenuId = line_bot_api.create_rich_menu(rich_menu=RichMenu.new_from_json_dict(menuJson))
print(lineRichMenuId) 
#拿到圖文選單id:richmenu-5a336562eb2511380bc3142ee63a1d5b
#>>>要好好保存,通常會保存到資料庫
```

```python
'''

將先前準備的菜單照片，以Post消息寄發給Line

    載入照片
    要求line_bot_api，將圖片傳到先前的圖文選單id

'''

#拿到圖文選單id:richmenu-5a336562eb2511380bc3142ee63a1d5b
#>>>要好好保存,通常會保存到資料庫

uploadImageFile=open("blockchain_demo.jpg",'rb') ###這個要事先上傳
# 讓line_bot_api上傳照片,並把生成的id 和圖文選單照片綁在一起
setImageResponse = line_bot_api.set_rich_menu_image(lineRichMenuId,'image/jpeg',uploadImageFile)

print(setImageResponse) #None代表成功
```

```python
'''

檢視用戶目前所綁定的菜單
    取出用戶id，並告知line_bot_api，
    line_bot_api傳回用戶所綁定的菜單
    印出

'''

#  https://api.line.me/v2/bot/user/{userId}/richmenu

# 查詢特定用戶 所使用的圖文選單
rich_menu_id = line_bot_api.get_rich_menu_id_of_user("Ud28913e45991f8c7531b94f85a9992a1")
print(rich_menu_id)
```

```python
'''

解除選單與特定用戶的綁定
    取出用戶id，並告知line_bot_api，
    line_bot_api解除用戶所綁定的菜單
'''
#解除用戶的圖文選單綁定
#參數:填寫要解除的用戶id
lineUnregisterUserMenuResponse=line_bot_api.unlink_rich_menu_from_user("Ud28913e45991f8c7531b94f85a9992a1")
print(lineUnregisterUserMenuResponse)
```
