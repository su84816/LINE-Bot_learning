```
'''
引用套件
'''

# 引用flask Web Server套件
from flask import Flask, request, abort, jsonify

# 載入json處理套件
import json

# 外部連結自動生成套件
from flask_ngrok import run_with_ngrok
```

```
'''
建置主程序
'''

# 設定Server啟用細節
#Flask 是做http server的套件
#如果要找得到照片,可以建立一個material的資料夾,把圖片放入 , 也可以放其他種檔案類型
#準備一個Flask應用,去本地端的 material抓圖片,當用戶訪問 xxx.io/material/1.jpg  他會去本地端的material資料夾內找 123.jpg >>>是在colab裡面建(因為現在程式是在colab上跑)
### material可以改任何名稱
app = Flask(__name__,static_url_path = "/material" , static_folder = "./material/")
#臨時生成一個http url,可以對外訪問,建議在開發時使用,生產環境上gcp並不需要
run_with_ngrok(app)
```

```
'''

為主程序創造多個API接口

想像思路
當外面的封包進來的時候，轉流到這個方法

'''
# 啟動主程序的API接口，最簡單版本
#用瀏覽器訪問 xxx.ngrok,xxx/ =>導引到此方法
#裝飾用詞,為了產出flask網址,和line機器人溝通
@app.route("/", methods=['GET'])
def callback():
     
    return 'Hello World'
```

```
# 第二隻接口，

# 啟動主程序的API接口，名為calculate，計算機功能，剖析v1參數與v2參數
#對 xxx.ngrok.xxx/calculate?v1=1&v2=2
@app.route("/calculate", methods=['GET'])
def calculate():
  #取用 query string內的v1與v2,並變成數字
    first_parameter = int(request.args.get('v1'))
    second_parameter = int(request.args.get('v2'))
    #作家減 回傳兩個數字做相加
    return str(first_parameter+second_parameter)
```

```
# 啟動主程序
#運行伺服器http application,app.run()會持續運行,之後要預載的方法,要寫在這裡上方
app.run()

# * Running on http://127.0.0.1:5000/ (Press CTRL+C to quit) >>> 這是本地端(訪問當前電腦,因為是colab所以我們不能訪問)
# * Running on http://5d16-35-204-88-156.ngrok.io >>> 這是外部訪問連結
```
